```python
import yfinance as yf
import matplotlib.pyplot as plt

# Specify the ticker symbols and the time period you want to retrieve data for
# tickers = ['IBM','MSFT','META','INTC','NEM','AU','AEM','GFI','AAPL','TSLA','PG','ZS','ADSK','QLYS']
tickers = ['IBM','MSFT','META','AAPL','TSLA','PG','ZS']
start_date = '2022-01-01'
end_date = '2022-11-29'

# Use a for loop to retrieve the stock data for each ticker
for ticker in tickers:
    stock_data = yf.Ticker(ticker).history(start=start_date, end=end_date)
    prices = stock_data['Close']
    
#     print(prices)
#     plt.plot(prices,linewidth=5, label=ticker, marker='o', markerfacecolor='black', markersize=5)
    plt.plot(prices,linewidth=5, label=ticker)

# Add a title and axis labels
plt.title('Closing Prices for Multiple Stocks')
plt.xlabel('Date')
plt.ylabel('Price')

# Add a legend
plt.legend(loc='upper right', fontsize = 25)

plt.rcParams["figure.figsize"] = (20,15)

plt.grid(True, color = "grey", linewidth = "1.4", linestyle = "-")

# display(plt.plot(x, y))

# plt.figure(figsize=(30, 6))
# display(plt.plot(x, y))
# Show the plot
plt.show()
```
